```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import multiprocessing
from IPython.display import display
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_curve, roc_auc_score
import xgboost as xgb



```

```python
# Load dataset
# df = pd.read_csv("/kaggle/input/weather-dataset-rattle-package/weatherAUS.csv")
file_path = "weatherAUS.csv"
df = pd.read_csv(file_path)

df
```

```python
# Display initial rows
display(df.head())

# Plot RainTomorrow distribution before processing
df['RainTomorrow'].value_counts().plot(kind="bar", title="Rain Tomorrow Distribution", rot=0)
plt.show()

# Count and visualize missing values
df.isna().sum().plot(kind="bar", title="Missing values", xlabel="Features", ylabel="Number of missing values")
plt.show()

# Drop rows with any missing values
df.dropna(inplace=True)

# Drop Date column
df.drop(["Date", "Location"], axis="columns", inplace=True)

display(df)
```

```python
# One-Hot Encoding for categorical variables
enc = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
ohe_cols = enc.fit_transform(df[["WindGustDir", "WindDir9am", "WindDir3pm", "RainToday"]])

# Concatenate encoded features and drop original categorical columns
df = pd.concat([df, ohe_cols], axis=1)
df.drop(["WindGustDir", "WindDir9am", "WindDir3pm", "RainToday"], axis=1, inplace=True)

# Convert target variable to binary
df['RainTomorrow'] = df['RainTomorrow'].replace({'No': 0, 'Yes': 1})

df
```

```python
# Split data into features (X) and target (y)
y = df['RainTomorrow']
X = df.drop(['RainTomorrow'], axis=1)

# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
# Standardize the data (important for PCA)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Perform PCA (without specifying components first to analyze variance)
pca = PCA()
pca.fit(X_train_scaled)

# Plot explained variance ratio
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(1, len(pca.explained_variance_ratio_) + 1), y=pca.explained_variance_ratio_, marker='o')
plt.title('Explained Variance Ratio of Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()

# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(1, len(pca.explained_variance_ratio_) + 1), 
             y=np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title('Cumulative Explained Variance of Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Explained Variance')
plt.show()

# Determine optimal number of components (keep enough to explain 90% variance)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
n_components = np.argmax(cumulative_variance >= 0.90) + 1
print(f"Number of components to keep: {n_components}")

# Perform PCA with the optimal number of components
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Convert PCA results to DataFrame
pca_df = pd.DataFrame(X_train_pca, columns=[f'PC{i+1}' for i in range(n_components)])

# Plot the first two principal components (scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_df['PC1'], y=pca_df['PC2'], hue=y_train, palette='viridis')
plt.title('PCA: First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Get CPU count for parallel processing
n_cpus = multiprocessing.cpu_count()

### SVM Model with GridSearchCV ###
parameters = {'C': [0.5, 1, 5, 10]}
svc = SVC(kernel='rbf', probability=True, random_state=0)
clf = GridSearchCV(svc, parameters, n_jobs=n_cpus)
clf.fit(X_train_pca, y_train)

# Display GridSearch results
results_df = pd.DataFrame(clf.cv_results_).sort_values(by=["rank_test_score"])
display(results_df[["params", "rank_test_score", "mean_test_score", "std_test_score"]])

# Train best SVM model
svc = SVC(C=clf.best_params_['C'], kernel='rbf', probability=True, random_state=0)
svc.fit(X_train_pca, y_train)
svc_y_pred_prob = svc.predict_proba(X_test_pca)[:,1]
svc_y_pred = svc.predict(X_test_pca)

# Evaluate SVM model
print('SVM Accuracy:', accuracy_score(y_test, svc_y_pred))
print('SVM Classification Report:\n', classification_report(y_test, svc_y_pred))

# Confusion matrix
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, svc_y_pred), display_labels=svc.classes_).plot()
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, svc_y_pred_prob)
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0,1], [0,1], 'k--')
plt.title('SVM ROC Curve')
plt.show()
print('SVM ROC AUC:', roc_auc_score(y_test, svc_y_pred_prob))


```

```python
### Random Forest with GridSearchCV ###
parameters = {'criterion': ['gini', 'entropy'], 'n_estimators': [100, 200, 300]}
random_forest = RandomForestClassifier(random_state=0)
clf = GridSearchCV(random_forest, parameters, n_jobs=n_cpus)
clf.fit(X_train_pca, y_train)

# Display results
results_df = pd.DataFrame(clf.cv_results_).sort_values(by=["rank_test_score"])
display(results_df[["params", "rank_test_score", "mean_test_score", "std_test_score"]])

# Train best RF model
random_forest = RandomForestClassifier(n_estimators=clf.best_params_['n_estimators'], criterion=clf.best_params_['criterion'], random_state=0)
random_forest.fit(X_train_pca, y_train)
rf_y_pred_prob = random_forest.predict_proba(X_test_pca)[:,1]
rf_y_pred = random_forest.predict(X_test_pca)

# Evaluate RF model
print('Random Forest Accuracy:', accuracy_score(y_test, rf_y_pred))
print('Random Forest Classification Report:\n', classification_report(y_test, rf_y_pred))

# Confusion matrix
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, rf_y_pred), display_labels=random_forest.classes_).plot()
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, rf_y_pred_prob)
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0,1], [0,1], 'k--')
plt.title('Random Forest ROC Curve')
plt.show()
print('Random Forest ROC AUC:', roc_auc_score(y_test, rf_y_pred_prob))

```

```python
### AdaBoost with GridSearchCV ###
parameters = {'learning_rate': [0.5, 1, 1.5, 2, 5], 'n_estimators': [100, 200, 300]}
adaboost = AdaBoostClassifier(algorithm='SAMME', random_state=0)
clf = GridSearchCV(adaboost, parameters, n_jobs=n_cpus)
clf.fit(X_train_pca, y_train)

# Display results
results_df = pd.DataFrame(clf.cv_results_).sort_values(by=["rank_test_score"])
display(results_df[["params", "rank_test_score", "mean_test_score", "std_test_score"]])

# Train best AdaBoost model
adaboost = AdaBoostClassifier(learning_rate=clf.best_params_['learning_rate'], n_estimators=clf.best_params_['n_estimators'], algorithm='SAMME', random_state=0)
adaboost.fit(X_train_pca, y_train)
adaboost_y_pred_prob = adaboost.predict_proba(X_test_pca)[:,1]
adaboost_y_pred = adaboost.predict(X_test_pca)

# Evaluate AdaBoost model
print('AdaBoost Accuracy:', accuracy_score(y_test, adaboost_y_pred))
print('AdaBoost Classification Report:\n', classification_report(y_test, adaboost_y_pred))

# Confusion matrix
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, adaboost_y_pred), display_labels=adaboost.classes_).plot()
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, adaboost_y_pred_prob)
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0,1], [0,1], 'k--')
plt.title('AdaBoost ROC Curve')
plt.show()
print('AdaBoost ROC AUC:', roc_auc_score(y_test, adaboost_y_pred_prob))

```

```python
### XGBoost with GridSearchCV ###
parameters = {'learning_rate': [0.01, 0.1], 'n_estimators': [100, 300, 500, 1000], 'max_depth': [3, 5, 7]}
xgboost = xgb.XGBClassifier(objective='binary:logistic', random_state=0)

clf = GridSearchCV(xgboost, parameters, n_jobs=n_cpus)
clf.fit(X_train_pca, y_train)

# Display results
results_df = pd.DataFrame(clf.cv_results_).sort_values(by=["rank_test_score"])
display(results_df[["params", "rank_test_score", "mean_test_score", "std_test_score"]])

# Train best XGBoost model
xgboost = xgb.XGBClassifier(learning_rate=clf.best_params_['learning_rate'], max_depth=clf.best_params_['max_depth'], n_estimators=clf.best_params_['n_estimators'], subsample=0.8, colsample_bytree=0.9, reg_alpha=0.1, reg_lambda=0.1, objective='binary:logistic', random_state=0)
xgboost.fit(X_train_pca, y_train)

# Evaluation
xgboost_y_pred_prob = xgboost.predict_proba(X_test_pca)[:,1]
xgboost_y_pred = xgboost.predict(X_test_pca)

print('XGBoost Accuracy:', accuracy_score(y_test, xgboost_y_pred))
print('XGBoost Classification Report:\n', classification_report(y_test, xgboost_y_pred))

# Confusion Matrix
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, xgboost_y_pred), display_labels=xgboost.classes_).plot()
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, xgboost_y_pred_prob)
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0,1], [0,1], 'k--')
plt.title('XGBoost ROC Curve')
plt.show()
print('XGBoost ROC AUC:', roc_auc_score(y_test, xgboost_y_pred_prob))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score

# Store metrics for each model
metrics = {
    "SVM": [
        accuracy_score(y_test, svc_y_pred),
        precision_score(y_test, svc_y_pred),
        recall_score(y_test, svc_y_pred),
        f1_score(y_test, svc_y_pred),
        roc_auc_score(y_test, svc_y_pred_prob),
    ],
    "Random Forest": [
        accuracy_score(y_test, rf_y_pred),
        precision_score(y_test, rf_y_pred),
        recall_score(y_test, rf_y_pred),
        f1_score(y_test, rf_y_pred),
        roc_auc_score(y_test, rf_y_pred_prob),
    ],
    "AdaBoost": [
        accuracy_score(y_test, adaboost_y_pred),
        precision_score(y_test, adaboost_y_pred),
        recall_score(y_test, adaboost_y_pred),
        f1_score(y_test, adaboost_y_pred),
        roc_auc_score(y_test, adaboost_y_pred_prob),
    ],
    "XGBoost": [
        accuracy_score(y_test, xgboost_y_pred),
        precision_score(y_test, xgboost_y_pred),
        recall_score(y_test, xgboost_y_pred),
        f1_score(y_test, xgboost_y_pred),
        roc_auc_score(y_test, xgboost_y_pred_prob),
    ],
}

# Define metric labels
metric_labels = ["Accuracy", "Precision", "Recall", "F1-Score", "AUC-ROC"]

# Define colors for each metric
colors = ["blue", "green", "red", "purple", "orange"]

# Number of models and metrics
num_models = len(metrics)
num_metrics = len(metric_labels)

# Set bar width
bar_width = 0.15
x = np.arange(num_models)  # X locations for models

# Create the plot
plt.figure(figsize=(10, 6))

for i, metric in enumerate(metric_labels):
    values = [metrics[model][i] for model in metrics]
    plt.bar(x + i * bar_width, values, width=bar_width, label=metric, color=colors[i])

# Add labels and title
plt.xlabel("Models")
plt.ylabel("Scores")
plt.title("Comparison of Model Performance Metrics")
plt.xticks(x + (num_metrics / 2 - 0.5) * bar_width, metrics.keys())  # Adjust x-ticks to be centered
plt.ylim(0, 1)  # Assuming all scores range between 0 and 1
plt.legend()
plt.show()
```
